```python
memory_size.value_counts()
```

```python
def toMB(x):
    if 'GB' in x :
        return float(x[:x.find('GB')])*1000
    else:
        return float(x[:x.find('TB')])*1000000
```

```python
memory_size['memory_size'] = memory_size['memory_size'].apply(toMB)
```

```python
data = data.join(memory_size)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split 
from sklearn.ensemble import RandomForestRegressor
```

```python
select_data.columns
```

```python
x,y = select_data.drop("Price",axis=1),select_data["Price"]
x_train , x_test,y_train,y_test = train_test_split(x,y,test_size = 0.15)
Scaler = StandardScaler()
x_train_scaled = Scaler.fit_transform(x_train)
x_test_scaled = Scaler.transform(x_test)
```

```python
forest = RandomForestRegressor()
forest.fit(x_train_scaled,y_train)
```

```python
x_train
```

```python
x_train_scaled[0]
```

```python
forest.score(x_test_scaled,y_test)
```

```python
y_predict = forest.predict(x_test_scaled)
```

```python
plt.figure(figsize=(12,8))
plt.scatter(y_predict,y_test)
plt.plot(range(0,500000),range(0,500000),c='black')
```
